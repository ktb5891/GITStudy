### 1변량 데이터의 관리

### 평균값(기댓값)

```python
N = len(fish_data)
sum_value = sp.sum(fish_data)
mu = sum_value/N
mu
```

```python
sp.mean(fish_data)
```

### 표본분산

```python
sigma_2_sample = sp.sum((fish_data-mu)**2)/N
sigma_2_sample
```

```python
fish_data
```

```python
fish_data-mu
```

```python
(fish_data-mu)**2
```

```python
sp.sum((fish_data-mu)**2)
```

#### ddof = 0 은 (N-0)으로 계산

```python
sp.var(fish_data, ddof = 0)
```

### 불편분산

```python
sigma_2 = sp.sum((fish_data-mu)**2)/(N-1)
sigma_2
```

#### ddof = 1 은 (N-1)로 계산

```python
sp.var(fish_data, ddof=1)
```

### 표준편차

```python
sigma = sp.sqrt(sigma_2)
sigma
```

```python
sp.std(fish_data, ddof = 1)
```

### 표준화

```python
fish_data-mu
```

```python
sp.mean(fish_data-mu)
```

```python
fish_data/sigma
```

```python
sp.std(fish_data/sigma, ddof=1)
```

```python
standard = (fish_data - mu)/sigma
standard
```

```python
sp.mean(standard) # 0과 다름없이 한없이 작은 값
```

```python
sp.std(standard, ddof = 1)
```

### 공분산(실습)

```python
cov_data = pd.read_csv("dataset/3-2-3-cov.csv")
cov_data
```

#### 데이터 분리

```python
x = cov_data["x"]
y = cov_data["y"]
```

#### 표본 크기

```python
N = len(cov_data)
```

#### 평균값 계산

```python
mu_x = sp.mean(x)
mu_y = sp.mean(y)
```

#### 공분산 계산 (N으로 설정)

```python
cov_sample = sum((x-mu_x)*(y-mu_y))/N
cov_sample
```

#### 공분산 계산 ((N-1)로 설정)

```python
cov = sum((x-mu_x)*(y-mu_y))/(N-1)
cov
```

#### 분산-공분산 행렬 (실습)

```python
sp.cov(x,y,ddof=0)
```

```python
sp.cov(x,y,ddof = 1)
```

### 피어슨 상관계수 (N-1로 계산)

#### 분산 계산

```python
sigma_2_x = sp.var(x,ddof = 1)
sigma_2_y = sp.var(y, ddof = 1)
```

#### 상관 계수

```python
rho = cov/sp.sqrt(sigma_2_x*sigma_2_y)
rho
```

### 피어슨 상관계수 (N으로 계산)

#### 분산 계산

```python
sigma_2_x_sample = sp.var(x,ddof=0)
sigma_2_y_sample = sp.var(y, ddof =0)
```

#### 상관 계수

```python
cov_sample/sp.sqrt(sigma_2_x_sample*sigma_2_y_sample)
```

#### -> N이나 N-1이나 상관계수는 같은 값을 가진다는 것을 알 수 있음

#### scipy의 corrcoef함수로 상관행렬 계산

```python
sp.corrcoef(x,y)
```
